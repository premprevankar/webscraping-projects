```python
import requests
from bs4 import BeautifulSoup
```

## User input

```python
title = input("Enter the freedom fighter name: ").strip()
wiki_link = "https://en.wikipedia.org/wiki/" + title.replace(" ", "_")
print("Wikipedia URL:", wiki_link)
```

```python
headers = {
    "User-Agent": "MyLearningBot/1.0 (contact: your_email@example.com)"
}

response = requests.get(wiki_link, headers=headers, timeout=10)

if response.status_code != 200:
    raise Exception("Wikipedia page not found")

soup = BeautifulSoup(response.text, "html.parser")
```

## Page title

```python
heading = soup.find("span", class_="mw-page-title-main").text
print("Title:", heading)
```

## Extract paragraphs

```python
para = ""
for p in soup.find_all("p"):
    para += p.get_text().replace("\n", " ")
```

## Remove citation numbers [1], [2]...

```python
for i in range(1, 500):
    para = para.replace(f"[{i}]", "")
```

```python
filename = heading.replace("/", "_") + ".txt"
with open(filename, "w", encoding="utf-8") as f:
    f.write(para)
```

## Save file

```python

print("Saved:", filename)
```
