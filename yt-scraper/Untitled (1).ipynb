## Imports

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

## Launch Browser & Open Channel

```python
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/@rajshamani/videos")
time.sleep(5)
```

## Parse Page Source

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
sp = soup.find("ytd-rich-grid-renderer")
```

## Extract Video Details

```python
data = []
for details in sp.find_all("ytd-rich-grid-media"):
    title_tag = details.find(
        "a",
        class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media"
    )
    title = title_tag.text.strip()
    vdo_link = "https://www.youtube.com" + title_tag.get("href")
    try:
        views = details.find(
            "span",
            class_="inline-metadata-item style-scope ytd-video-meta-block"
        ).text
    except:
        views = None
    try:
        date_time = details.find_all(
            "span",
            class_="inline-metadata-item style-scope ytd-video-meta-block"
        )[1].text
    except:
        date_time = None
    try:
        thumbnail_link = details.find("img").get("src").split("?")[0]
    except:
        thumbnail_link = None
    data.append((title, vdo_link, views, date_time, thumbnail_link))


   
```

## Create DataFrame

```python
data_set = pd.DataFrame(data,columns=['Title' , 'Vdo link', 'Views','Upload date ', 'Thumbnail lnk'])
```
